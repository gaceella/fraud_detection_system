```python
X = df.drop('Class', axis=1)
y = df['Class']
```

```python
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()

X_scaled = scaler.fit_transform(X)
```

```python
import pandas as pd

X_scaled = pd.DataFrame(X_scaled, columns=X.columns)
```

```python
X_scaled['amount_zscore'] = (
    X_scaled['Amount'] - X_scaled['Amount'].mean()
) / X_scaled['Amount'].std()
```

```python
X_scaled['time_diff'] = X_scaled['Time'].diff().fillna(0)
```

```python
X_scaled.shape
```

```python
sns.histplot(X_scaled['amount_zscore'], bins=50)

plt.title("Amount Z-Score Distribution")

plt.show()
```

```python
X_scaled.head()

X_scaled.shape
```

```python
from sklearn.ensemble import IsolationForest

# Convert dataframe to numpy array
X_array = X_scaled.values

iso = IsolationForest(
    n_estimators=100,
    contamination=0.01,
    random_state=42
)

iso_pred = iso.fit_predict(X_array)

# Convert predictions
iso_pred = [1 if x == -1 else 0 for x in iso_pred]
```

```python
sum(iso_pred)
```

```python
sum(y)
```

```python
from sklearn.metrics import classification_report

print(classification_report(y, iso_pred))
```

```python
import matplotlib.pyplot as plt

plt.hist(iso_pred, bins=2)

plt.title("Isolation Forest Predictions")

plt.xlabel("Prediction (0=Normal, 1=Anomaly)")
plt.ylabel("Number of Transactions")

plt.show()
```

```python
from sklearn.neighbors import LocalOutlierFactor

lof = LocalOutlierFactor(
    n_neighbors=20,
    contamination=0.01
)

lof_pred = lof.fit_predict(X_array)

lof_pred = [1 if x == -1 else 0 for x in lof_pred]
```

```python
sum(lof_pred)
```

```python
from sklearn.cluster import DBSCAN

dbscan = DBSCAN(
    eps=3,
    min_samples=5
)

db_pred = dbscan.fit_predict(X_array)

# Convert noise (-1) to fraud label
db_pred = [1 if x == -1 else 0 for x in db_pred]
```

```python
sum(db_pred)
```

```python
import numpy as np

fraud_score = np.array(iso_pred) + np.array(lof_pred) + np.array(db_pred)
```

```python
import matplotlib.pyplot as plt

plt.hist(fraud_score, bins=4, edgecolor='black')

plt.title("Hybrid Fraud Score Distribution")
plt.xlabel("Fraud Score")
plt.ylabel("Number of Transactions")

plt.show()
```

```python
y_pred = [1 if x >= 2 else 0 for x in fraud_score]
```

```python
from sklearn.metrics import classification_report

print(classification_report(y, y_pred))
```

```python
y_pred = [1 if x >= 3 else 0 for x in fraud_score]

from sklearn.metrics import classification_report
print(classification_report(y, y_pred))
```
